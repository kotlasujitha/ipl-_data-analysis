```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')

print(f"Matches: {matches.shape[0]} rows")
print(f"Deliveries: {deliveries.shape[0]} rows")
```

```python
wins = matches['winner'].value_counts()

plt.figure(figsize=(10, 5))
sns.barplot(x=wins.values, y=wins.index, palette='Blues_r', hue=wins.index, legend=False)
plt.title('IPL — Total wins by team')
plt.xlabel('Wins')
plt.tight_layout()
plt.savefig('charts/team_wins.png', dpi=150)
plt.show()
```

```python
top_bat = (deliveries.groupby('batsman')['batsman_runs']
           .sum()
           .sort_values(ascending=False)
           .head(10))

plt.figure(figsize=(10, 5))
sns.barplot(x=top_bat.values, y=top_bat.index, 
            palette='Oranges_r', hue=top_bat.index, legend=False)
plt.title('IPL — Top 10 all-time run scorers')
plt.xlabel('Total runs')
plt.tight_layout()
plt.savefig('charts/top_batsmen.png', dpi=150)
plt.show()
```

```python
print(deliveries.columns.tolist())
```

```python
top_bat = (deliveries.groupby('batter')['batsman_runs']
           .sum()
           .sort_values(ascending=False)
           .head(10))

plt.figure(figsize=(10, 5))
sns.barplot(x=top_bat.values, y=top_bat.index, 
            palette='Oranges_r', hue=top_bat.index, legend=False)
plt.title('IPL — Top 10 all-time run scorers')
plt.xlabel('Total runs')
plt.tight_layout()
plt.savefig('charts/top_batsmen.png', dpi=150)
plt.show()
```

```python
matches['toss_won_match'] = (matches['toss_winner'] == matches['winner'])
pct = matches['toss_won_match'].mean() * 100

print(f"Toss winner wins the match: {pct:.1f}% of the time")
print(f"\nToss decisions:\n{matches['toss_decision'].value_counts()}")

matches['toss_decision'].value_counts().plot(
    kind='pie', autopct='%1.1f%%',
    colors=['#378ADD','#1D9E75'],
    startangle=90
)
plt.title('Toss decision — bat vs field')
plt.ylabel('')
plt.savefig('charts/toss.png', dpi=150)
plt.show()
```

```python
season_runs = (
    deliveries.merge(matches[['id','season']], left_on='match_id', right_on='id')
    .groupby('season')['total_runs'].sum().reset_index()
)

plt.figure(figsize=(10, 4))
plt.plot(season_runs['season'], season_runs['total_runs'],
         marker='o', color='#378ADD', linewidth=2)
plt.fill_between(season_runs['season'], season_runs['total_runs'], 
                 alpha=0.1, color='#378ADD')
plt.title('Total runs scored per IPL season')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.tight_layout()
plt.savefig('charts/season_runs.png', dpi=150)
plt.show()
```
